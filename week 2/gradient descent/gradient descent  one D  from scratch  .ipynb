```python
from sklearn.datasets import make_regression
```

```python
#creating linear data sets 
x ,y = make_regression(n_samples= 4  , n_features= 1 , n_informative = 1 , n_targets = 1 , noise = 80, random_state = 8 )
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter( x , y )
```

```python
#applying ole  -. ordinary lest squared 
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x , y) 
```

```python
lr.intercept_  , lr.coef_
```

```python
plt.scatter(x , y ) 
plt.plot( x, lr.predict(x) , color = 'red')
```

```python
# so this is the best line that our model predict  okey 
```

```python
#step 1 . Asume slop is constant and euqal to 9.52634892  or 9.52 
# and lets assume inital value for b is zero 
#then we  know that  y = mx + b  by  putting the values we can calulate the y(hat) valus
# but the line made by this more incorrect and so predicted values
```

```python
ypred = ((9.52 * x) + 0 ).reshape(4)
```

```python
ypred
```

```python
import seaborn as sns
plt.scatter(x , y )
plt.plot(x , lr.predict(x)  , label = 'Ordinary lest square' )
plt.plot(x , ypred , color = 'red' , label = " b = 0  , x = 9.52(const) " ) 
plt.legend()
plt.show()
```

```python
# Lets take a splot at b = 0
import numpy as np
m = 9.52
b = 0 
#step 2 calculate the slope 
slope_of_b = -2 * np.sum(y - m*x - b )

#loss function ma 
# m lai constant manera  b ko terms ma derivative nikalera aako equation ho mathi ko -2 ... wala
```

```python
#now , take learning rate 
learning_rate  = 0.01 
# stepsize =slope * learning rate 
stepsize = slope_of_b * learning_rate
stepsize ## yeeti size step ( -- wa ++  hudai aagi bandxa baneko ) 
```

```python
b
```

```python
# calculating new step size 
b = b - stepsize 
b
```

```python
# we know our m is constant and we got our new b  
ypred1 = ((m*x) + b).reshape(4)
```

```python
plt.scatter(x , y) 
plt.plot(x ,lr.predict(x) , color = 'red' , label = "OLS"  ) 
plt.plot(x ,ypred , color = 'black' , label = "b = 0 ") 
plt.plot(x ,ypred1 , color = 'green' , label =f"b = {b}") 
plt.legend()

plt.show()
```

```python
# wee see most accurate line is red that is ols 
# second most accurate line is black 
# and last is b = 0 
```

```python
# now taking another iterator and again observaing result 
slope_b = -2 * np.sum( y - m*x.ravel() - b ) 
slope_b
```

```python
stepsize = slope_b *learning_rate
stepsize
```

```python
b = b - stepsize 
```

```python
ypred2 = (( m*x ) + b ).reshape(4)
```
