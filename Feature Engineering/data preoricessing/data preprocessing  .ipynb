```python
import pandas as pd
df = pd.DataFrame([
['green', 'M', 10.1, 'class2'],
['red', 'L', 13.5, 'class1'],
['blue', 'XL', 15.3, 'class2']])
df.columns = ['color', 'size', 'price', 'classlabel']
df
dfs = df.copy()
```

```python
inv_size_mapping = {v: k for k, v in size_mapping.items()}
df['size'] = df['size'].map(inv_size_mapping)
```

## Mininmizing dimension ( one hot enccoding using only most frequent variables) 

suppose yout have **40 Cars Brand** Out of which **10 Car Brands** have very low **frequency** while others have very hight 
frequency 

#### THen we can reduce the dimension by puttling all low frequent cars on new others category ### 


```python
df = pd.read_csv("cars.csv")
```

```python
counts = df['brand'].value_counts()
counts
```

```python
threshold = 100  #min 100 frequency 
rep = counts[counts < threshold].index
```

```python
rep # cars which has less than 100 frequency 
```

```python
pd.get_dummies(df['brand'].replace(rep , "others") , drop_first = True)
```
